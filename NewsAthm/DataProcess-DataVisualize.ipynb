```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取包含类别和语料数的 CSV 文件
file_path = "category_counts_all_new_Nums3000p_CategoryMerge_new_undersampled.csv"
data = pd.read_csv(file_path)

# 提取类别数量
category_counts = data['num']

# 绘制饼状图
plt.figure(figsize=(10, 8))
plt.pie(category_counts, labels=range(1, len(category_counts) + 1), autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Category Distribution by Index')
plt.show()
```
