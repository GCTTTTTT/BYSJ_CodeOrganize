```python
import pandas as pd
import numpy as np


# 读取CSV文件
df = pd.read_csv("./datasets/deduplicated_mangoNews_Nums3000p_CategoryMerge_new_undersampled.csv",low_memory=False,lineterminator="\n",nrows=1000)

# 将 NaN 替换为空字符串
df['abstract'] = df['abstract'].fillna('')
df['body'] = df['body'].fillna('')

# 统计abstract列内容是否是body列内容中的一部分的个数
is_subset_count = 0
not_subset_count = 0

for index, row in df.iterrows():
    abstract = str(row['abstract'])  # 将 abstract 列的内容转换为字符串
    body = str(row['body'])          # 将 body 列的内容转换为字符串
    if abstract in body:
        is_subset_count += 1
    else:
        not_subset_count += 1
        print(abstract)
        print("==============")
        print(body)
        print("==============")

print("Abstract 是 Body 的一部分的行数:", is_subset_count)
print("Abstract 不是 Body 的一部分的行数:", not_subset_count)
```
