```python
import pandas as pd
# 'deduplicated_mangoNews.csv'中类别统计并保存
# 读取CSV文件
# file_path = 'mangoNews_Example.csv'  # 测试版数据集（规模小5m）
# file_path = 'mangoNews.csv'          # 完整版数据集（13g） 需要分块读
# file_path = 'mangoNews_Example_100000.csv'          # 100000行数据集（2.2g）
# file_path = 'mangoNews_Example_10000.csv'          # 10000行数据集（190m)

# file_path = './deduplicated_mangoNews.csv'          # 去重后完整数据集（12.9g) 需要分块读
# file_path = 'deduplicated_mangoNews_Example_10000.csv'          # 去重后10000行数据集（190m)
# file_path = 'deduplicated_mangoNews_Example_100000.csv'          # 去重后100000行数据集（2.1g)

# file_path = './deduplicated_mangoNews_Nums3000p.csv'          # 去重后保留类别对应语料数在3000+的数据集（9.9g) 需要分块读
# file_path = './deduplicated_mangoNews_Nums3000p_CategoryMerge.csv'          # 去重后保留类别对应语料数在3000+的数据集（9.9g) 需要分块读
file_path = './deduplicated_mangoNews_Nums3000p_CategoryMerge_new.csv'          # 去重后保留类别对应语料数在3000+的数据集（9.9g) 需要分块读
file_path = './deduplicated_mangoNews_Nums3000p_CategoryMerge_new_undersampled.csv'        # 数据平衡后数据集 12095*12=145140条 1.2g 随机欠采样
file_path = './deduplicated_mangoNews_Nums3000p_CategoryMerge_new_randsample145140.csv'        # 数据平衡前数据集随机采样 145140条 1.2g 用于对比

# 分块读取：
# 读取CSV文件，使用chunksize进行分块读取
# file_path = 'your_dataset.csv'  # 替换为你的数据集路径
chunk_size = 100000  # 根据实际情况调整分块大小
data_chunks = pd.read_csv(file_path, usecols=['body', 'category1'], chunksize=chunk_size, low_memory=False,lineterminator="\n")

# 初始化空的DataFrame，用于存储统计结果
result_df = pd.DataFrame(columns=['category1', 'num'])

# 遍历分块数据
for chunk in data_chunks:
    # 统计 'category1' 列中每种类别的个数
    print(len(chunk))
    category_counts = chunk['category1'].value_counts().reset_index()
    category_counts.columns = ['category1', 'num']

    # 合并到总的结果DataFrame
    result_df = pd.concat([result_df, category_counts], ignore_index=True)

# 合并结果中的重复项，得到最终的统计结果
final_result = result_df.groupby('category1')['num'].sum().reset_index()

# 按照 'num' 列降序排序
final_result = final_result.sort_values(by='num', ascending=False)

# print(final_result)

# 将最终结果保存到CSV文件
final_result.to_csv('category_counts_all_new_Nums3000p_CategoryMerge_new_randsample145140.csv', index=False)

# 打印结果
# 设置显示选项，完整输出结果
pd.set_option('display.max_rows', None)
print("Category Counts (Sorted):")
print(final_result)
# 恢复默认显示选项
pd.reset_option('display.max_rows')
```
