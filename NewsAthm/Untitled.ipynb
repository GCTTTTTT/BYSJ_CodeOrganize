```python
import pandas as pd

file_path = './mangoNews.csv'
chunk_size = 100

# 使用迭代器和逐行读取的方式读取文件的前100行
chunks = pd.read_csv(file_path, chunksize=chunk_size)

# 逐块处理数据
for i, chunk in enumerate(chunks):
    if i * chunk_size >= 100:
        break  # 读取前100行后停止

    # 格式化输出前100行数据
    print(chunk)
```

```python
# 读取CSV文件的前100行
import pandas as pd

file_path = './mangoNews.csv'  # 替换为你的CSV文件路径
# data = pd.read_csv(file_path, low_memory=False)
chunk_size = 100000  # 设置适当的块大小

# 逐块读取CSV文件
chunks = pd.read_csv(file_path, chunksize=chunk_size, low_memory=False)

cnt_c1 = 0
cnt_c2 = 0
cnt_rows = 0
for chunk in chunks:
    # 在这里对每个块的数据进行处理
    # ...
    nan_count_category1 = chunk['category1'].isna().sum()
    nan_count_category2 = chunk['category2'].isna().sum()
    cnt_c1+= nan_count_category1
    cnt_c2+= nan_count_category2
    cnt_rows+=len(chunk)

print(cnt_c1)
print(cnt_c2)
print(cnt_rows)

# print(nan_count_category2)
# num_rows = len(data)
# print("Number of rows in the CSV file:", num_rows)
# # 打印前100行的'category1'和'category2'列的内容
# print("Category 1:")
# print(data['category1'])

# print("\nCategory 2:")
# print(data['category2'])
# data['category1'].to_string()
```
