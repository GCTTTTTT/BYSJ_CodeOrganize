```python
import matplotlib.pyplot as plt
import networkx as nx
from graphviz import Digraph

# 创建有向图
dot = Digraph(comment='mBert-Bi-LSTM-Self-Attention Model')

# 添加输入层节点
dot.node('input', 'Input Text', shape='rectangle')

# 添加BERT编码层节点
dot.node('bert', 'BERT Encoder', shape='rectangle')
dot.edge('input', 'bert', label='input_ids, attention_mask')

# 添加Bi-LSTM层节点
dot.node('bilstm', 'Bi-LSTM', shape='rectangle')
dot.edge('bert', 'bilstm', label='last_hidden_state')

# 添加Self-Attention层节点
dot.node('attention', 'Self-Attention', shape='rectangle')
dot.edge('bilstm', 'attention', label='lstm_outputs')

# 添加输出层节点
dot.node('output', 'Output Layer', shape='rectangle')
dot.edge('attention', 'output', label='attention_outputs')

# 添加分类输出节点
dot.node('classification', 'Classification', shape='circle')
dot.edge('output', 'classification', label='logits')

# 渲染并保存图形
dot.render('mBert-Bi-LSTM-Self-Attention', view=True)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# 创建一个新的图形
fig, ax = plt.subplots(figsize=(12, 8))

# 绘制输入层
input_text = patches.Rectangle((0.5, 7.5), 2, 0.5, linewidth=1, edgecolor='black', facecolor='lightblue')
ax.add_patch(input_text)
ax.text(1.5, 7.7, 'Input Text', ha='center', va='center')

# 绘制BERT编码层
bert_encoder = patches.Rectangle((0.5, 6), 2, 1, linewidth=1, edgecolor='black', facecolor='lightgreen')
ax.add_patch(bert_encoder)
ax.text(1.5, 6.5, 'BERT Encoder', ha='center', va='center')

# 绘制Bi-LSTM层
bilstm = patches.Rectangle((0.5, 4.5), 2, 1, linewidth=1, edgecolor='black', facecolor='lightyellow')
ax.add_patch(bilstm)
ax.text(1.5, 5, 'Bi-LSTM', ha='center', va='center')

# 绘制Self-Attention层
self_attention = patches.Rectangle((0.5, 3), 2, 1, linewidth=1, edgecolor='black', facecolor='lightpink')
ax.add_patch(self_attention)
ax.text(1.5, 3.5, 'Self-Attention', ha='center', va='center')

# 绘制输出层
output_layer = patches.Rectangle((0.5, 1.5), 2, 1, linewidth=1, edgecolor='black', facecolor='lightcyan')
ax.add_patch(output_layer)
ax.text(1.5, 2, 'Output Layer', ha='center', va='center')

# 绘制分类输出
classification = patches.Circle((1.5, 0.5), 0.3, linewidth=1, edgecolor='black', facecolor='white')
ax.add_patch(classification)
ax.text(1.5, 0.5, 'Classification', ha='center', va='center')

# 绘制箭头连接
ax.arrow(1.5, 7.5, 0, -0.4, head_width=0.1, head_length=0.2, linewidth=1, color='black')
ax.text(2.6, 7.1, 'input_ids\nattention_mask', ha='left', va='center')

ax.arrow(1.5, 6, 0, -0.4, head_width=0.1, head_length=0.2, linewidth=1, color='black')
ax.text(2.6, 5.6, 'last_hidden_state', ha='left', va='center')

ax.arrow(1.5, 4.5, 0, -0.4, head_width=0.1, head_length=0.2, linewidth=1, color='black')
ax.text(2.6, 4.1, 'lstm_outputs', ha='left', va='center')

ax.arrow(1.5, 3, 0, -0.4, head_width=0.1, head_length=0.2, linewidth=1, color='black')
ax.text(2.6, 2.6, 'attention_outputs', ha='left', va='center')

ax.arrow(1.5, 1.5, 0, -0.7, head_width=0.1, head_length=0.2, linewidth=1, color='black')
ax.text(2.6, 1.1, 'logits', ha='left', va='center')

# 设置图形属性
ax.set_xlim(0, 5)
ax.set_ylim(0, 8.5)
ax.axis('off')

# 显示图形
plt.tight_layout()
plt.show()
```
